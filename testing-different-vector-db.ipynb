```python
import os

folder_path = '/kaggle/input/processed-math-book/testv1'
md_files = [f for f in os.listdir(folder_path) if f.endswith('.md')]

all_content = {}
for file_name in md_files:
    with open(os.path.join(folder_path, file_name), 'r', encoding='utf-8') as f:
        all_content[file_name] = f.read()
```

```python
all_topics = []
for chapter_name, content in all_content.items():
    topics = content.split('*****')
    for topic in topics:
        topic = topic.strip()
        if topic:
            # Optionally, you can remove the chapter name from the start of each topic
            # This is helpful for cleaner embeddings
            # topic_text = topic.replace(chapter_name.replace('.md', ''), '').strip()
            all_topics.append(topic)
```

```python
# Load a pre-trained model
model = SentenceTransformer('all-MiniLM-L6-v2') 
```

```python
import numpy as np
import os
from sentence_transformers import SentenceTransformer
import faiss
```

```python
# 2. Initialize the FAISS index
# We need to get the dimension of the embeddings first
embedding_dimension = model.get_sentence_embedding_dimension()
index = faiss.IndexFlatL2(embedding_dimension) # Using L2 distance
```

```python
# 3. Create a list to store the original text chunks
all_chunks_text = []

print("Starting chunk-by-chunk embedding and indexing...")
for i, chunk in enumerate(all_topics):
    # Embed the current chunk
    chunk_vector = model.encode([chunk])
    
    # Ensure the vector is in the correct format (float32)
    chunk_vector = np.float32(chunk_vector)
    
    # Add the vector to the FAISS index
    index.add(chunk_vector)
    
    # Store the original text chunk in our new list
    all_chunks_text.append(chunk)

    # Optional: Print progress
    if (i + 1) % 100 == 0:
        print(f"Processed {i + 1}/{len(all_topics)} chunks.")

print("\nChunk-by-chunk embedding and indexing complete.")
print(f"Total number of vectors in the index: {index.ntotal}")
```

```python
# Example of a query
query = "বাস্তব সংখ্যা কি ?"
query_vector = model.encode([query])
distances, ids = index.search(np.float32(query_vector), k=5)

print("\nTop 5 results:")
for i, faiss_id in enumerate(ids[0]):
    retrieved_chunk = all_chunks_text[faiss_id]
    print(f"Rank {i+1}:")
    print(f"Distance: {distances[0][i]:.4f}")
    print(f"Chunk: {retrieved_chunk}\n")
```

```python
from qdrant_client import QdrantClient, models
```

```python
# Initialize the Qdrant client in in-memory mode
client = QdrantClient(":memory:")
```

```python
# Define a collection name
collection_name = "math_topics"

# Check if the collection already exists and delete it to start fresh
if client.collection_exists(collection_name):
    client.delete_collection(collection_name)
```

```python
embedding_dimension = model.get_sentence_embedding_dimension()
client.create_collection(
    collection_name=collection_name,
    vectors_config=models.VectorParams(size=embedding_dimension, distance=models.Distance.COSINE),
)
```

```python
# Start chunk-by-chunk embedding and indexing
print("Starting chunk-by-chunk embedding and indexing...")

# You can use a list to store both the vector and payload for easier batch upload
points_to_upload = []

for i, chunk in enumerate(all_topics):
    # Embed the current chunk
    chunk_vector = model.encode(chunk).tolist() # Qdrant expects a list
    
    # Qdrant can store the chunk text directly as a 'payload' with the vector
    points_to_upload.append(
        models.PointStruct(
            id=i,  # Assign a unique ID to each point
            vector=chunk_vector,
            payload={"text": chunk} # Store the original text as a payload
        )
    )

    # Optional: Print progress
    if (i + 1) % 100 == 0:
        print(f"Processed {i + 1}/{len(all_topics)} chunks.")

# Upload all the points in one go for efficiency
client.upload_points(
    collection_name=collection_name,
    points=points_to_upload
)

print("\nChunk-by-chunk embedding and indexing complete.")
print(f"Total number of vectors in the index: {client.count(collection_name).count}")
```

**Semantic Search**

```python
# Example of a query
query = "বাস্তব সংখ্যা কি ?"
query_vector = model.encode(query).tolist()

search_results = client.search(
    collection_name=collection_name,
    query_vector=query_vector,
    limit=5, # Number of results to return
    with_payload=True # Include the original text in the results
)

print("\nTop 5 search results:")
for result in search_results:
    print(f"Score: {result.score:.4f}")
    print(f"Chunk: {result.payload['text']}\n")
```

**Hybride search**

```python
# Sample a query that benefits from both semantic and keyword search
query = "বাস্তব সংখ্যা কি?"
```

```python
query_vector = model.encode(query).tolist()
semantic_results = client.search(
    collection_name=collection_name,
    query_vector=query_vector,
    limit=10,
    with_payload=True
)
```

```python
keyword_results = client.scroll(
    collection_name=collection_name,
    scroll_filter=models.Filter(
        must=[
            models.FieldCondition(
                key="text",
                match=models.MatchText(text=query)
            )
        ]
    ),
    limit=10,
    with_payload=True,
    # This is a key difference: `scroll` returns a tuple of (points, next_page_offset)
    # We only care about the points for this case.
)[0] 
```

```python
# --- Part 3: Combine Results using Reciprocal Rank Fusion (RRF) ---
def rrf(results_lists, k=60):
    fused_scores = {}
    for results in results_lists:
        for rank, result in enumerate(results):
            doc_id = result.id
            score = 1.0 / (k + rank + 1)
            if doc_id not in fused_scores:
                fused_scores[doc_id] = 0.0
            fused_scores[doc_id] += score
    
    reranked_results = sorted(fused_scores.items(), key=lambda item: item[1], reverse=True)
    return reranked_results
```

```python
# Get a list of the search results from each method
all_results = [semantic_results, keyword_results]

# Perform RRF to get a combined, ranked list of document IDs
reranked_ids = rrf(all_results)

# Now, retrieve the documents corresponding to the top-ranked IDs
print("Top 5 Hybrid Search Results:")
for i, (doc_id, score) in enumerate(reranked_ids[:5]):
    point = client.retrieve(
        collection_name=collection_name,
        ids=[doc_id],
        with_payload=True,
    )[0]
    print(f"Rank {i+1}: (Score: {score:.4f})")
    print(f"Chunk: {point.payload['text']}\n")
```
